```python
df.columns = ['InvestHorizon','WhereMoneyNow','ReturnExp','Age','InvestableAsset','Income','MaxDrawD']
```

```python
from sklearn.preprocessing import OneHotEncoder
cat = df.columns[:-1]
X = df[cat]

enc = OneHotEncoder(handle_unknown='ignore')
newx = enc.fit_transform(X).toarray()

enc_df = pd.DataFrame(newx)
```
